```python
for column in ['education', 'self_employed']:
    if column in df.columns:  # Kiểm tra cột có tồn tại không
        insert_loc = df.columns.get_loc(column)
        df = pd.concat([
            df.iloc[:, :insert_loc], 
            pd.get_dummies(df[[column]]),  
            df.iloc[:, insert_loc+1:] if insert_loc+1 < df.shape[1] else df.iloc[:, insert_loc:]
        ], axis=1)
```

```python
X1 = df.drop('loan_status', axis=1)  # Loại bỏ cột 'loan_status' khỏi df để tạo tập đặc trưng
y1 = df['loan_status']  # Lấy cột 'loan_status' làm biến mục tiêu
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
from imblearn.combine import SMOTETomek
from imblearn.over_sampling import ADASYN
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler

#adasyn = ADASYN(sampling_strategy='minority', random_state=42)
#smote_tomek = SMOTETomek(random_state=42)
#smote_enn = SMOTEENN(random_state=42)
smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X1, y1)

y_sm.value_counts()
```

```python
#Tách dữ liệu và tạo model
```

```python
# Chọn cột Input, không sử dụng Customer ID vì nó không hữu hiệu, và Output là loan_status
# Tạo X bao gồm tất cả các cột trừ 'loan_id' và no_of_dependents', 'loan_status'
X = df.drop(columns=['loan_id', 'no_of_dependents', 'loan_status'])
```

```python
# Tạo y bao gồm cột 'loan_status'
y = df['loan_status']
# Split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)
```

```python
# Show X_train
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
# Cột cần scale
scale_columns = ['cibil_score', 'loan_amount','loan_term','residential_assets_value','commercial_assets_value',	'luxury_assets_value',	'bank_asset_value',	'income_annum']
scaler = StandardScaler()
X_train[scale_columns] = scaler.fit_transform(X_train[scale_columns])
X_test[scale_columns] = scaler.transform(X_test[scale_columns])
```

```python
#Logistics Regression
```

```python
from sklearn.linear_model import LogisticRegression

# Khởi tạo mô hình Logistic Regression mà không dùng class_weight
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score,recall_score
# Đánh giá mô hình
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("F1 Score:", f1)
print("Precision:", precision)
print("Recall:", recall)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
# In classification report
print(classification_report(y_test, y_pred))
```

```python
#Mô hình Logistic Regression đã đạt:
#Accuracy ~ 90%
#Precision, Recall, F1-score trung bình đều ~ 0.89 – 0.90 trở lên.
#Các chỉ số này cho thấy mô hình phân loại tốt trên cả hai lớp, ít bị thiên vị cho lớp đa số hoặc thiểu số.
#Đặc biệt, lớp 1 (thường là lớp quan trọng nếu đó là lớp “dương tính” trong bài toán) có Precision = 0.92 và Recall = 0.91, tức là mô hình bắt được rất nhiều mẫu thuộc lớp 1 và ít sai sót.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Giả sử best_lr_classifier đã được huấn luyện và X_train là DataFrame chứa các đặc trưng
# Với mô hình phân loại nhị phân, best_lr_classifier.coef_ là mảng 2 chiều có shape (1, n_features)
# Lấy các hệ số (coefficients) của mô hình và tính giá trị tuyệt đối
feature_importance = np.abs(model.coef_[0])

# Chuẩn hóa độ quan trọng về thang điểm 0-100%
feature_importance = 100.0 * (feature_importance / feature_importance.max())

# Sắp xếp theo thứ tự tăng dần của độ quan trọng
sorted_idx = np.argsort(feature_importance)
pos = np.arange(len(feature_importance))

# Vẽ biểu đồ thanh ngang
plt.figure(figsize=(8, 6))
plt.barh(pos, feature_importance[sorted_idx], align='center', color='skyblue')
plt.yticks(pos, np.array(X_train.columns)[sorted_idx], fontsize=10)
plt.xlabel('Feature Importance (%)')
plt.title('Feature Importance của Logistic Regression')
plt.show()
```
